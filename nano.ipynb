```python
from pathlib import Path
from pypdf import PdfReader

book_paths = list(Path("books").glob("*.pdf"))

all_text = ""

for path in book_paths:
    reader = PdfReader(path)
    for page in reader.pages:
        page_text = page.extract_text()
        if page_text:
            all_text += page_text + "\n"

print(len(all_text))
print(all_text[:500])
```

```python
chars = sorted(list(set(all_text)))
vocab_size = len(chars)
print(chars)
len(chars)
```

```python
stoi = {ch:i for i,ch in enumerate(chars)}
itos = {i:ch for i,ch in enumerate(chars)}
```

```python
def encode(s):
    return [stoi[ch] for ch in s]

def decode(ids):
    return ''.join(itos[i] for i in ids)
```

```python
word = "hello"
print(encode(word))
print(decode(encode(word)))
```

```python
import torch
torch.manual_seed(1337)

data = torch.tensor(encode(all_text), dtype=torch.long)

batch_size = 128
block_size = 64

n = int(0.9 * len(data))

train_data = data[:n]
val_data = data[n:]

def get_batch(split):
    #to generate small batch of data of inputs x and target y
    source = train_data if split=="train" else val_data
    ix = torch.randint(len(data) - block_size, (batch_size,))
    x = torch.stack([source[i:i+block_size] for i in ix])
    y = torch.stack([source[i+1:batch_size+1] for i in ix])
    return x,y
```
