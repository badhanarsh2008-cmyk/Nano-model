```python
all_text = ''

data = PdfReader("books/Sherlock holmes.pdf")
text = ''
for page in data.pages:
    text += page.extract_text()
all_text += text 
print(len(text))
print(f"length of all text {len(all_text)}")
print(text[:100])
```
